# Brief History

* The name "Pandas" has a reference to both "Panel Data", and "Python Data Analysis" and was created by Wes McKinney in 2008. It has function to analyse, clean, explore and manipulate the data.
* It helps to work with large datasets and find the results based on statistical theories.
* Pandas can clean the messy datasets, and make them readable and relevant.
* Pandas can also delete the rows that are not relevant, or containing wrong values, like empty or NULL values. This is called the cleaning the data.

* Pandas is a labeled, columnar, Numpy-backed relational engine with a sophisticated block-oriented memory manager.

* It is a BlockManager that stores heterogeneous data in homogeneous blocks for cache efficiency.

# pandas Data Structures

Data structures in pandas: Series and DataFrame

### Series: It is a one dimensional array-like object containing a sequence of values of the same type and an associated array of data labels, called its index.

```python
import pandas as pd

obj = pd.Series([4, 7, -5, 3])
obj
```

```python
print(obj.array)
print(obj.index) #pandas has default indexing start with 0 otherwise mentioned
```

```python
obj2 = pd.Series([4, 7, -5, 3], index=['a','b','c','d'])
obj2
```

* Another way to think about the Series as a fixed-length, ordered dictionary as it represents a mapping of index values to the data values.

* We can override the key values in the series values using dictionary keys.


```python
dict_data = {"x": 35000, "y": 71000, "z": 16000, "u": 5000}
obj3 = pd.Series(dict_data)
obj3
```

```python
new_ind = ["a","b","x","y"]
dict_data = {"x": 35000, "y": 71000, "z": 16000, "u": 5000}

obj4 = pd.Series(dict_data, index = new_ind)
# Here it will replace the two values found from the original dictionary but appear as NaN(Not a Number) for other values.
obj4
```

```python
print(obj3)
obj3.index = ["a","b","c","d"]
# Index alteration
print(obj3)
```

```python
import pandas as pd
import numpy as np

s = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'], name='my_series', dtype='int32')

print("--- Series Object ---")
print(s)

print("\n--- Internal Attributes ---")
print("s.values          →", type(s.values), s.values.dtype)          # NumPy array (or Arrow in some cases)
print("s.index           →", type(s.index), s.index.dtype)
print("s.name            →", s.name)
print("s.dtype           →", s.dtype)
print("s.shape           →", s.shape)
print("s.ndim            →", s.ndim)
print("s._values.base    →", s._values.base is None)   # memory view info

# ExtensionArray view (Pandas 2.x+)
print("\n--- ExtensionArray (new unified interface) ---")
print("s.array           →", type(s.array))   # IntegerArray, StringArray, etc.
```

# DataFrame

* It is a rectangular table of data and contains an ordered, named collection of columns, each of which can be a different value type(numeric, string, Boolean etc.)

* It is a two dimensional labeled data structure and stores multiple homogeneous blocks smartly.

* It is a dict-like container of Series object that share the exact same index(axis = 0), with a separate container Index on axis = 1.

* It is actually not the python dict of series, but a "BlockManager" that stores data in homogeneous Numpy blocks for maximum cache efficiency and vectorized operations.


```python
import numpy as np
import pandas as pd

df = pd.DataFrame({
    'int_col':    [1, 2, 3],
    'float_col':  [4.1, 5.2, 6.3],
    'str_col':    ['a', 'b', 'c'],
    'cat_col':    pd.Categorical(['x', 'y', 'x'])
})

print("---- DataFrame -----")
print(df)

print("\n--- The Real Storage: BlockManager ---")
print(df._mgr)   # Shows how block manager converts all the dict containers into numpy blocks.

print("\n--- Blocks Breakdown ---")
for i, blk in enumerate(df._mgr.blocks):
    print(f"Block {i}: {type(blk).__name__} | dtype={blk.dtype} | shape={blk.shape} | "
          f"columns={df.columns[blk.mgr_locs.indexer]}")
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': np.arange(5, dtype='int32'),
    'B': np.random.randn(5),
    'C': pd.Categorical(['x','y','x','y','x'])
})

print("Number of blocks:", len(df._mgr.blocks))

for i, blk in enumerate(df._mgr.blocks):
    arr = blk.values
    addr = hex(arr.ctypes.data) if hasattr(arr, 'ctypes') else "Arrow buffer"
    print(f"Block {i} ({blk.dtype}): address = {addr}")  # It shows that the how blocks are the stored in the memory.
```
