# DataFrame

* It is a rectangular table of data and contains an ordered, named collection of columns, each of which can be a different value type(numeric, string, Boolean etc.)

* It is a two dimensional labeled data structure and stores multiple homogeneous blocks smartly.

* It is a dict-like container of Series object that share the exact same index(axis = 0), with a separate container Index on axis = 1.

* It is actually not the python dict of series, but a "BlockManager" that stores data in homogeneous Numpy blocks for maximum cache efficiency and vectorized operations.


````python
import numpy as np
import pandas as pd

df = pd.DataFrame({
    'int_col':    [1, 2, 3],
    'float_col':  [4.1, 5.2, 6.3],
    'str_col':    ['a', 'b', 'c'],
    'cat_col':    pd.Categorical(['x', 'y', 'x'])
})

print("---- DataFrame -----")
print(df)

print("\n--- The Real Storage: BlockManager ---")
print(df._mgr)   # Shows how block manager converts all the dict containers into numpy blocks.

print("\n--- Blocks Breakdown ---")
for i, blk in enumerate(df._mgr.blocks):
    print(f"Block {i}: {type(blk).__name__} | dtype={blk.dtype} | shape={blk.shape} | "
          f"columns={df.columns[blk.mgr_locs.indexer]}")
````

````python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': np.arange(5, dtype='int32'),
    'B': np.random.randn(5),
    'C': pd.Categorical(['x','y','x','y','x'])
})

print("Number of blocks:", len(df._mgr.blocks))

for i, blk in enumerate(df._mgr.blocks):
    arr = blk.values
    addr = hex(arr.ctypes.data) if hasattr(arr, 'ctypes') else "Arrow buffer"
    print(f"Block {i} ({blk.dtype}): address = {addr}")  # It shows that the how blocks are the stored in the memory.
````

### Possible data inputs to the DataFrame constructor

| Type | Notes |
| :--- | :--- |
| **2D ndarray** | A matrix of data, passing optional row and column labels |
| **Dictionary of arrays, lists, or tuples** | Each sequence becomes a column in the DataFrame; all sequences must be the same length |
| **NumPy structured/record array** | Treated as the “dictionary of arrays” case |
| **Dictionary of Series** | Each value becomes a column; indexes from each Series are unioned together to form the result’s row index if no explicit index is passed |
| **Dictionary of dictionaries** | Each inner dictionary becomes a column; keys are unioned to form the row index as in the “dictionary of Series” case |
| **List of dictionaries or Series** | Each item becomes a row in the DataFrame; unions of dictionary keys or Series indexes become the DataFrame’s column labels |
| **List of lists or tuples** | Treated as the “2D ndarray” case |
| **Another DataFrame** | The DataFrame’s indexes are used unless different ones are passed |
| **NumPy MaskedArray** | Like the “2D ndarray” case except masked values are missing in the DataFrame result |

* Ranking in pandas is the process of assigning a numerical score to each element in a dataset based on its position relative to others.

1. The df.rank() method
Parameters:
* axis: Rank across rows(axis = 0) or columns(axis = 1)
* ascending: If false, the largest value gets the rank 1 (best for top 10 list)
* pct: If true, it returns the rank as percentile.

2. Handling ties:

When two values are identical then pandas provide following steps:

| Method | Description | Example (Values: 10, 20, 20, 30) |
| :--- | :--- | :--- |
| `average` | (Default) Assigns the average of the ranks that would have been assigned. | 1, 2.5, 2.5, 4 |
| `min` | Uses the lowest rank for the whole group (Competition ranking). | 1, 2, 2, 4 |
| `max` | Uses the highest rank for the whole group. | 1, 3, 3, 4 |
| `first` | Assigns ranks in the order they appear in the data. | 1, 2, 3, 4 |
| `dense` | Like `min`, but the next rank after a tie is the next integer (no gaps). | 1, 2, 2, 3 |

3. Ranking with Groupby:

```python
df['Group_Rank'] = df.groupby('Classroom')['Score'].rank(ascending=False)
```



````python
import pandas as pd

data = {'Sales': [100, 250, 250, 500, 50]}
df = pd.DataFrame(data)

# Rank 1 being the highest seller using 'dense' to avoid skipping numbers
df.rank(ascending=False, method = 'dense')
````

# Unique Values, Value Counts, and Membership

1. unique() and nunique():
- unique(): Returns an array of the unique values in the order they appear. It does not sort them.
- nunique(): Returns the number of unique values. By default, it ignores NaN values (use dropna=False to include them).

2. Frequency Counts: value_counts(): It counts how many times each unique value appears.
* By default, the result is sorted by frequency (highest first)
* Set normalize=True to get the relative frequencies (percentages) instead of raw counts

3. Membership: isin(): The isin() method is a highly efficient way to check if values in a Series or DataFrame exist within a provided list, set, or another Series. It returns a boolean mask (True/False).

4. The "Index" Connection: Index.get_indexer(): If you have a list of items and want to know their integer positions in another index, you use get_indexer.
```python
# Match unique values to a reference list to get their positions
obj = pd.Series(['c', 'a', 'b', 'b', 'c', 'a'])
uniques = pd.Series(['c', 'b', 'a'])

pd.Index(uniques).get_indexer(obj)
# Output: array([0, 2, 1, 1, 0, 2])
```
