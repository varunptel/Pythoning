# Brief History

* The name "Pandas" has a reference to both "Panel Data", and "Python Data Analysis" and was created by Wes McKinney in 2008. It has function to analyse, clean, explore and manipulate the data.
* It helps to work with large datasets and find the results based on statistical theories.
* Pandas can clean the messy datasets, and make them readable and relevant.
* Pandas can also delete the rows that are not relevant, or containing wrong values, like empty or NULL values. This is called the cleaning the data.

* Pandas is a labeled, columnar, Numpy-backed relational engine with a sophisticated block-oriented memory manager.

* It is a BlockManager that stores heterogeneous data in homogeneous blocks for cache efficiency.

# pandas Data Structures

Data structures in pandas: Series and DataFrame

### Series: It is a one dimensional array-like object containing a sequence of values of the same type and an associated array of data labels, called its index.

```python
import pandas as pd

obj = pd.Series([4, 7, -5, 3])
obj
```

```python
print(obj.array)
print(obj.index) #pandas has default indexing start with 0 otherwise mentioned
```

```python
obj2 = pd.Series([4, 7, -5, 3], index=['a','b','c','d'])
obj2
```

* Another way to think about the Series as a fixed-length, ordered dictionary as it represents a mapping of index values to the data values.

* We can override the key values in the series values using dictionary keys.


```python
dict_data = {"x": 35000, "y": 71000, "z": 16000, "u": 5000}
obj3 = pd.Series(dict_data)
obj3
```

```python
new_ind = ["a","b","x","y"]
dict_data = {"x": 35000, "y": 71000, "z": 16000, "u": 5000}

obj4 = pd.Series(dict_data, index = new_ind)
# Here it will replace the two values found from the original dictionary but appear as NaN(Not a Number) for other values.
obj4
```

```python
print(obj3)
obj3.index = ["a","b","c","d"]
# Index alteration
print(obj3)
```
