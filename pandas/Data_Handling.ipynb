# Data Clearning and Preparations using pandas

### Handling Missing Data:
For data with float64 dtype, pandas uses the floating-point value NaN (Not a Number) to represent missing data.


```python
import numpy as np
import pandas as pd

float_data = pd.Series([1.2, -3.5, np.nan, 0])
print(float_data)
print(float_data.isna())
```

### Filtering Out missing Data

| Method | Description |
| :--- | :--- |
| `dropna` | Filter axis labels based on whether values for each label have missing data, with varying thresholds for how much missing data to tolerate. |
| `fillna` | Fill in missing data with some value or using an interpolation method such as `"ffill"` or `"bfill"`. |
| `isna` | Return Boolean values indicating which values are missing/NA. |
| `notna` | Negation of `isna`; returns `True` for non-NA values and `False` for NA values. |



```python
data = pd.Series([1,np.nan, 0, np.nan, 7])
print(data)
print(data.dropna())
```

```python
df = pd.DataFrame([[1., 6.5, 3.], [1., np.nan, np.nan],
                   [np.nan, np.nan, np.nan], [np.nan, 6.5, 3.]])

print(df)
print(df.dropna(how="all"))  # how="all" will drop only rows that are all NA
```

```python
# Lets add column 4 with all NA values
df[4] = np.nan
df
```

```python
df.dropna(axis = 'columns', how = 'all')
```

```python
df = pd.DataFrame(np.random.standard_normal((7,3)))
df.iloc[:4, 1] = np.nan
df.iloc[:2, 2] = np.nan
df
```

```python
df.dropna()
```

```python
df.dropna(thresh = 2)
```

### Filling in Missing Data

fillna function arguments:
| Argument | Description |
| :--- | :--- |
| **value** | Scalar value or dictionary-like object to use to fill missing values. |
| **method** | Interpolation method: one of `"bfill"` (backward fill) or `"ffill"` (forward fill); default is `None`. |
| **axis** | Axis to fill on (`"index"` or `"columns"`); default is `axis="index"`. |
| **limit** | For forward and backward filling, maximum number of consecutive periods to fill. |

```python
# Filling in missing data

df.fillna(0) # replaces the NA values with the scalar value 0
```

```python
# calling fillna with dictionary, we can replace different value for different columns

df.fillna({1: 0.5, 2: 0})
```

```python
df = pd.DataFrame(np.random.standard_normal((6, 3)))
df.iloc[2:, 1] = np.nan
df.iloc[4:, 2] = np.nan
df
```

```python
print(df.fillna(method = 'ffill'))
print(df.fillna(method = 'ffill', limit = 2))
```

```python
# We can do a lot of things with the fillna() method

data = pd.Series([1, np.nan, 3.5, np.nan, 7])

data = data.fillna(data.mean())
data
```

### Data Transformation

There are several methods linked to the data transformations

1) Removing Duplicates:
* drop_duplicates() returns a DataFrame with rows where the duplicated array is False filtered out.
* These methods by default considers all the columns, we can specify the subset to them to detect the duplicates.

```python
import pandas as pd
import numpy as np

data = pd.DataFrame({'k1':['one','two'] * 3 + ['two'],
                     'k2':[1,1,2,3,3,4,4]})
print(data)
print(data.duplicated()) # Returns boolean value for duplicated rows
```

```python
print(data.drop_duplicates())
```

```python
data['v1'] = range(7)
print(data)
print(data.drop_duplicates(subset = ['k1'])) # duplicated values are dropped in k1 column
print(data.drop_duplicates(subset = ['k1', 'k2'], keep = 'last')) # duplicated values are dropped in k1 and k2 columns while keeping the last value instead of the first
```

2. Transforming Data Using a Function or Mapping
* The map function on a series accepts the function of dictionary like key-value pair of values to do the transformation.


```python
data = pd.DataFrame({"food": ["bacon", "pulled pork", "bacon",
                              "pastrami", "corned beef", "bacon",
                              "pastrami", "honey ham", "nova lox"],
                     "ounces": [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

```python
meat_to_animal = {
 "bacon": "pig",
 "pulled pork": "pig",
 "pastrami": "cow",
 "corned beef": "cow",
 "honey ham": "pig",
 "nova lox": "salmon"
} # mapping function
```

```python
data['animal'] = data['food'].map(meat_to_animal)
data
```

3. Replacing Values

```python
import pandas as pd
import numpy as np

data = pd.Series([1, -999., 2, -999., -1000, 3])
data
```

```python
data.replace(-999, np.nan) # replacing -999 with NA values
```

```python
data.replace([-999, -1000], [np.nan, 0]) # Replacing multiple values
data.replace({-999: np.nan, -1000: 0})  # THis is same thing as above but with dict like object
```

3. Renaming the Axis Indexes

* One of the way to rename the axis by mapping a function onto the original values.
* If we want transformed version without modifying the dataset is by using the method of rename()



```python
import numpy as np
import pandas as pd

df = pd.DataFrame(np.arange(12).reshape(3,4),
                  index = ['Ohio', 'Colorado', 'New York'],
                  columns = ['one', 'two', 'three', 'four'])

df
```

```python
def transform(x):
  return x[:4].upper()

df.index.map(transform)
```

```python
df.index = df.index.map(transform)
df
```

```python
df.rename(index = str.title, columns = str.upper)
```

```python
df.rename(index = {'OHIO': "INDIANA"},
          columns = {'three': 'This'}
          )
```
