```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3], [4, 5, 6], marker='D') # This uses the Diamond marker
plt.show()
```

## Marker Size and Marker Color:

* Use ms parameter inside plot function for marker size.

* Use mec for marker color.

* Markerfacecolor or the shorter mfc to set the color inside the edge of the markers:

* You can also use Hexadecimal color values or any of the 140 supported color names. https://matplotlib.org/stable/users/explain/colors/colors.html



```python
y = np.array([1,5,6,8,42,0,512,2,54,1,22,666])


plt.plot(y, 's--g', ms = 4, mec = 'r', mfc = 'r')
```

# Matplotlib Line

---

### Linestyle:
we can use the keyword argument linestle (or ls) to change the style of the plotted line.

| Style Name (linstyle) | Syntax (ls) | Description |
| :--- | :---: | :--- |
| `'solid'` (default) | `'-'` | Solid line |
| `'dotted'` | `':'` | Dotted line |
| `'dashed'` | `'--'` | Dashed line |
| `'dashdot'` | `'-.'` | Dashed/dotted line |
| `'None'` or `' '` | `''` | No line (invisible) |

---

### Line Width:
We can use linewidth or lw to change the width of the line.

---

### Multiple Line:
You can plot as many lines as you like by simply adding more plt.plot() functions. You can also plot many lines by adding the points for the x- and y-axis for each line in the same plt.plot() function.



```python
import numpy as np
import matplotlib.pyplot as plt

y1 = np.random.rand(10)
y2 = np.random.rand(10)

plt.plot(y1,marker = "D" ,linestyle = "dashed")
plt.plot(y2,'o', ls = ":", color = 'red')
```

```python
x1 = np.random.randint(10, size = 10)
x2 = np.random.randint(10, size = 10)

plt.plot(x1, ls = '-.', lw = '2.2')
plt.plot(x2, ls = '-.', lw = '2.2')
```

# Matplotlib Labels and Title

* In PyPlot, we can use xlabel() and ylabel() functions to set a label for the x- and y-axis.
* With Pyplot, you can use the title() function to set a title for the plot.
* You can use the fontdict parameter in xlabel(), ylabel(), and title() to set font properties for the title and labels.
* You can use the loc parameter in title() to position the title.
* Legal values are: 'left', 'right', and 'center'. Default value is 'center'.

https://www.w3schools.com/python/matplotlib_labels.asp


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.plot(x, y)

plt.title("Sports Watch Data", fontdict = font1, loc = 'left')
plt.xlabel("Average Pulse", fontdict = font1)
plt.ylabel("Calorie Burnage", fontdict = font1)

plt.show()
```
