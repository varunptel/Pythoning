# Matplotlib Labels and Title

* In PyPlot, we can use xlabel() and ylabel() functions to set a label for the x- and y-axis.
* With Pyplot, you can use the title() function to set a title for the plot.
* You can use the fontdict parameter in xlabel(), ylabel(), and title() to set font properties for the title and labels.
* You can use the loc parameter in title() to position the title.
* Legal values are: 'left', 'right', and 'center'. Default value is 'center'.

https://www.w3schools.com/python/matplotlib_labels.asp


````python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.plot(x, y)

plt.title("Sports Watch Data", fontdict = font1, loc = 'left')
plt.xlabel("Average Pulse", fontdict = font1)
plt.ylabel("Calorie Burnage", fontdict = font1)

plt.show()
````

# Matplotlib Subplot

### Subplot is a classic way to a grid of multiple plots inside a single figure window.

plt.subplot(nrows, ncols, index, **kwargs)

* nrows: number of rows in the grid
* ncols: number of columns in the grid
* index: which subplot you want (index start with 1 not "0")
* **kwargs: optional settings (e.g., projection='polar', facecolor=...)

## Indexing:
The index fills the grid left-to-right, then top-to-bottom
image.png



````python
import numpy as np
import matplotlib.pyplot as plx

x = np.linspace(1,10,50)

plt.subplot(2,1,1)  # Third argument here represent the order. This will be first one.
plt.plot(x, np.sin(x), 'b-')
plt.title("Subplot(2,1,1): Sin(x)")
plt.grid()

plt.subplot(2,1,2) # This will be second one.
plt.plot(x, np.cos(x), 'r-')
plt.title("Subplot(2,1,2): Cos(x)")
plt.grid()

plt.suptitle("Title for the entire graph")
plt.tight_layout() # This prevents the title and the labels from overlapping
plt.show()
````

# Matplotlib Scatter Plots

It plots individual points(x,y) as markers, making it perfect for showing relationships, correlations, clusters, or distributions between two variables.

### Basic Syntax:
```
plt.scatter(x, y, s=None, c=None, marker='o', cmap=None, alpha=None,
            edgecolors=None, linewidths=None, **kwargs)
```

* x,y: Array coordinates
* s: marker size
* c: marker color
* marker: shape ('o', 's', '^', '.', etc.)
* cmap: colormap when c is numeric
* alpha: transparency(0=invisible, 1=solid)
* edgecolor/linewidths = outline controller

````python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
x = np.random.rand(100) * 10
y = np.cos(x)

plt.scatter(x,y, marker = 'o', alpha = 0.7)
plt.title("Basic Scatter Plot")
plt.xlabel("x values")
plt.ylabel("y Values")
plt.grid(True)
````

````python
color = np.random.rand(100)
sizes = 50 * np.random.rand(100) + 20

sc = plt.scatter(x , y, s= sizes, c = color, cmap = 'plasma', alpha = 0.8, edgecolor = 'w' , linewidth = 0.5)
plt.colorbar(sc, label = 'Color Scale')
plt.title('Scatter with Size & Color')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.grid()

plt.show()
````

# Matplotlib Bar Plot
It is perfect for comparing quantities across categories. Unlike scatter plots (which show relationships between two continuous variables), bar plots emphasize discrete categories and their associated values.

### Basic Syntax
```
plt.bar(x, height, width=0.8, bottom=None, align='center',
        color=None, edgecolor=None, linewidth=None, **kwargs)
```

* X: categories (strings, numbers, or positions)
* height: bar height
* width: bar width



````python
# Single value bar plot

import numpy as np
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 12, 67, 34]


plt.bar(categories, values, color = 'blue', edgecolor = 'black', linewidth = 2)
plt.title("Basic Bar Graph")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.grid(True, alpha = 0.2)
plt.tight_layout()
````

````python
# Group values bar plot

import numpy as np
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D', 'E']
x = np.arange(len(categories))
width = 0.35
values1 = [23, 45, 12, 67, 34]
values2 = [30, 35, 20, 50, 40]

plt.figure(figsize = (10, 6 ))
plt.bar(x - width/2, values1, width, label='Group 1', color='skyblue', edgecolor='navy')
plt.bar(x + width/2, values2, width, label='Group 2', color='lightcoral', edgecolor='darkred')
plt.title('Grouped Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(axis = 'y', alpha = 0.2)
````

# Matplotlib Histograms

The go-to visualization for showing the distribution of a single numerical variable.

### Basic Syntax:

```
plt.hist(x, bins=None, range=None, density=False, weights=None,
         cumulative=False, bottom=None, histtype='bar', align='mid',
         orientation='vertical', color=None, edgecolor=None, **kwargs)
```

* x : array of values.
* bins: number of bins to represent the histogram
* density = true: Normalizes so that the total area = 1.
* histtype: 'bar'(default), 'barstacked', 'step', 'stepfilled'
* orientation: orientation of the chart horizontal or vertical



````python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.normal(loc = 0, scale = 1, size = 10000)

plt.hist(x, bins = 50,density = True, color = 'blue', edgecolor = 'navy', alpha = 0.8)
plt.title("Basic Histogram")
plt.xlabel("Values")
plt.ylabel("Probability distribution")
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize = (8,5))

plt.hist(x, bins = 20, density = True, cumulative= False,
         color = 'green', edgecolor = 'black', alpha = 0.85, label = 'Density')
plt.hist(x, bins = 20, density = True, cumulative= True,
         color = 'green', edgecolor = 'black', alpha = 0.85, label = 'Cumulative')

plt.title('Density + Cumulative Histogram')
plt.xlabel('Value')
plt.ylabel('Probability / Cumulative Probability')
plt.show()

````

````python
group1 = np.random.normal(0, 1, 8000)
group2 = np.random.normal(2, 1.5, 8000)

plt.figure(figsize=(9, 5))
plt.hist(group1, bins=25, alpha=0.7, label='Group 1 (μ=0)', color='skyblue', edgecolor='navy')
plt.hist(group2, bins=25, alpha=0.7, label='Group 2 (μ=2)', color='lightcoral', edgecolor='darkred')
plt.title('Overlaid Histograms – Comparing Distributions')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
````

````python
# 2D Histogram of x(independent variable), and y(dependent variable)
import numpy as np
import matplotlib.pyplot as plt

x = np.random.normal(0, 1, 2000)
y = x * 0.5 + np.random.normal(0,1,2000)

plt.hist2d(x, y, bins = 50, cmap = 'plasma')
plt.colorbar(label = 'Count')
plt.title('2D Histogram')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
````
